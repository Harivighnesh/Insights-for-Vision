```python
def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = dataframe[(dataframe[column] < lower) | (dataframe[column] > upper)]
    return outliers

print("Outlier detection:")
numerical_columns = df.select_dtypes(include=np.number).columns

for col in numerical_columns:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col}: {len(outliers)} outliers")
```

```python
def replace_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df


for col in df.select_dtypes(include=np.number).columns:
    df = replace_outliers_iqr(df, col)
```

```python
for col in numerical_columns:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col}: {len(outliers)} outliers")
```

# Modifying wrong data (number of friends are -1,-2.) with Median Value

```python
median_outdoor=df[df['Time spent playing outdoor sports'] >= 0]['Time spent playing outdoor sports'].median()
df.loc[df['Time spent playing outdoor sports'] < 0, 'Time spent playing outdoor sports'] =median_outdoor
```

```python
median_friends=df[df['Number of friends'] >= 0]['Number of friends'].median()
df.loc[df['Number of friends'] < 0, 'Number of friends'] =median_friends
```
